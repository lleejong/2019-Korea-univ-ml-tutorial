```
from gensim.models import Word2Vec
import logging
```

```
def main():
    # import modules & set up logging
    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

    # read sentences, Iterate over a file that contains sentences
    sentences = Word2Vec.LineSentence('./sample_data/wiki_en_split.txt')

    # train word2vec, sg=1 : skip-gram // workers : parallelization
    model = Word2Vec(sentences, window=5, sg=1, size=300, workers=4)

    # model save
    model.save("./sample_data/wiki_en_model")
```

```
def model_example():
  
  try:
    model = Word2Vec.load('./sample_data/wiki_en_model')
    vector = model.wv['apple']
    print(vector)
  except KeyError:
    print("This word not in vocabulary")
  
  try:
    target_word = ['computer']
    print(model.wv.most_similar(positive=target_word, topn=6))
   
  except KeyError:
    print("This word not in vocabulary")
  
  
```

```
if __name__ == "__main__":
    main()
#     model_example()
```
